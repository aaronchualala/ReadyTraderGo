## Relevant data
Time (arr1[0]), ETFPirce (arr1[1]), FuturePrice (arr1[2])

```python
def getRawData(raw1): # gets the relevant information of the dataframe
    timeArr = raw1["Time"].to_numpy()
    etfArr = raw1["EtfPrice"].to_numpy()
    futArr = raw1["FuturePrice"].to_numpy()
    return np.array([timeArr, etfArr, futArr])

def displayRawData(raw1, x): # x is the ID of the dataset
    compressedData = getRawData(raw1)
    timeArr, etfArr, futArr = compressedData[0], compressedData[1], compressedData[2]
    fig, ax = plt.subplots(figsize=(24,6))
    ax.plot(timeArr, futArr, color="blue", lw=0.5, label="Future")
    ax.plot(timeArr, etfArr, color="red", lw=0.5, label="ETF")
    ax.set_title("Data Set " + x)
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Traded price ($)")
    ax.legend()
    plt.show()
```

```python
fig, ax = plt.subplots(nrows = 4, figsize = (24, 24))

compressedData = getRawData(raw1)
timeArr, etfArr, futArr = compressedData[0], compressedData[1], compressedData[2]
etfFutRatioArr = etfArr / futArr
etfFutRatioMean = np.nanmean(etfFutRatioArr)
print("Mean ratio for dataset 1:", etfFutRatioMean)
etfFutRatioVar = np.nanvar(etfFutRatioArr)
print("Mean var for dataset 1:", etfFutRatioVar)
ax[0].plot(timeArr, etfFutRatioArr, lw=1)
ax[0].axhline(y=etfFutRatioMean, color="red")
ax[0].set_title("Data Set 1")
ax[0].set_xlabel("Time (s)")
ax[0].set_ylabel("Ratio of ETF to Future price")

compressedData = getRawData(raw2)
timeArr, etfArr, futArr = compressedData[0], compressedData[1], compressedData[2]
etfFutRatioArr = etfArr / futArr
etfFutRatioMean = np.nanmean(etfFutRatioArr)
print("Mean ratio for dataset 2:", etfFutRatioMean)
etfFutRatioVar = np.nanvar(etfFutRatioArr)
print("Mean var for dataset 2:", etfFutRatioVar)
ax[1].plot(timeArr, etfFutRatioArr, lw=1)
ax[1].axhline(y=etfFutRatioMean, color="red")
ax[1].set_title("Data Set 2")
ax[1].set_xlabel("Time (s)")
ax[1].set_ylabel("Ratio of ETF to Future price")

compressedData = getRawData(raw3)
timeArr, etfArr, futArr = compressedData[0], compressedData[1], compressedData[2]
etfFutRatioArr = etfArr / futArr
etfFutRatioMean = np.nanmean(etfFutRatioArr)
print("Mean ratio for dataset 3:", etfFutRatioMean)
ax[2].plot(timeArr, etfFutRatioArr, lw=1)
etfFutRatioVar = np.nanvar(etfFutRatioArr)
print("Mean var for dataset 3:", etfFutRatioVar)
ax[2].axhline(y=etfFutRatioMean, color="red")
ax[2].set_title("Data Set 3")
ax[2].set_xlabel("Time (s)")
ax[2].set_ylabel("Ratio of ETF to Future price")

compressedData = getRawData(raw4)
timeArr, etfArr, futArr = compressedData[0], compressedData[1], compressedData[2]
etfFutRatioArr = etfArr / futArr
etfFutRatioMean = np.nanmean(etfFutRatioArr)
print("Mean ratio for dataset 4:", etfFutRatioMean)
etfFutRatioVar = np.nanvar(etfFutRatioArr)
print("Mean var for dataset 4:", etfFutRatioVar)
ax[3].plot(timeArr, etfFutRatioArr, lw=1)
ax[3].axhline(y=etfFutRatioMean, color="red")
ax[3].set_title("Data Set 4")
ax[3].set_xlabel("Time (s)")
ax[3].set_ylabel("Ratio of ETF to Future price")

plt.show()
```
